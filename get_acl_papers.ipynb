# 下载ACL论文

```python
import os
from tqdm import tqdm
```

```python
base_url = "https://www.aclweb.org/anthology/P18-1"

for i in tqdm(range(257)):
    if i < 10 :
        j = "00" + str(i)
    elif i < 100 :
        j = "0" + str(i)
    else:
        j = str(i)
    url = base_url + j
#     print(url)
    cmd = "wget " + url
    os.system(cmd)
```
